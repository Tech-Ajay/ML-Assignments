##Scenario 1: Parking Lot Occupancy (Binomial Distribution)


```python
import numpy as np
import matplotlib.pyplot as plt

# Number of simulations
num_simulations = 10000

# Probability of a car occupying a spot
prob_car = 0.7

# Simulate parking lot occupancy
occupancy = np.random.binomial(50, prob_car, num_simulations)

# Calculate expected occupancy and standard deviation
expected_occupancy = np.mean(occupancy)
std_dev_occupancy = np.std(occupancy)

# Plot the distribution of parking lot occupancy
plt.hist(occupancy, bins=range(0, 51), density=True, alpha=0.6)
plt.xlabel("Number of Cars in Parking Lot")
plt.ylabel("Probability")
plt.title("Parking Lot Occupancy Distribution")
plt.grid(True)
plt.show()

print(f"Expected Occupancy: {expected_occupancy:.2f}")
print(f"Standard Deviation of Occupancy: {std_dev_occupancy:.2f}")
```

##Scenario 2: Bike Sharing Station Arrivals (Poisson Distribution)

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of simulations
num_simulations = 10000

# Arrival rate of bikes per hour
arrival_rate = 1

# Simulate bike arrivals in a 2-hour period
arrivals = np.random.poisson(arrival_rate * 2, num_simulations)

# Calculate expected arrivals and standard deviation
expected_arrivals = np.mean(arrivals)
std_dev_arrivals = np.std(arrivals)

# Plot the distribution of bike arrivals
plt.hist(arrivals, bins=range(0, max(arrivals) + 1), density=True, alpha=0.6)
plt.xlabel("Number of Bike Arrivals in 2 Hours")
plt.ylabel("Probability")
plt.title("Bike Arrival Distribution")
plt.grid(True)
plt.show()

print(f"Expected Arrivals: {expected_arrivals:.2f}")
print(f"Standard Deviation of Arrivals: {std_dev_arrivals:.2f}")
```
